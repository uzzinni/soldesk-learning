```python
fig = plt.figure()

ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

import numpy as np
ax1.plot(np.random.standard_normal(50).cumsum(), color='black', linestyle='dashed')
ax2.plot(np.random.standard_normal(30).cumsum(), color='black', linestyle='dashed', marker='_')
ax3.hist(np.random.standard_normal(100))
ax4.scatter(np.arange(30), np.arange(30) + 3 * np.random.standard_normal(30))
```
