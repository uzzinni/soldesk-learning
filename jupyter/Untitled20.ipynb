```python
# 이산화
# pd.cut 메소드 사용       298p
import pandas as pd

ages = [20,22,25,27,21,23,37, 31,61,45,41,32]

bins = [18,25,35,60,100]

age_cate = pd.cut(ages, bins)
age_cate.codes
```

```python
pd.value_counts(age_cate)
```
